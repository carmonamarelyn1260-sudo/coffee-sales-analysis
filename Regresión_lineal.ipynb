```python
modelo = linear_model.LinearRegression().fit(datos[['Promedio']],datos[['Tiempo']])
```

```python
#son los estimdores
modelo.intercept_,modelo.coef_
```

```python
#Sacar promedio de x
mean_x = datos['Promedio'].mean()
mean_x
```

```python
#Sacar promedio de y
mean_y = datos['Tiempo'].mean()
mean_y
```

```python
#Saca los x - xbarra
datos['x-xbarra'] = datos['Promedio']-mean_x
#Saca los y -ybarra
datos['y-ybarra'] = datos['Tiempo']-mean_y
```

```python
  #sum (xi-xbarra)(yi-yabrra)/n
datos['(x-xbarra)(y-ybarra)'] = datos['x-xbarra']*datos['y-ybarra']
```

```python
covarianza = datos['(x-xbarra)(y-ybarra)'].sum()/len(datos)
covarianza
```

```python
#desviacion estandar
sdx =np.std(datos['Promedio'])
sdy =np.std(datos['Tiempo'])
```

```python
#Sacar el coeficiente de correlación
p = covarianza/(sdx*sdy)
p
```

```python
datos['Y_hat'] = modelo.predict(datos[['Promedio']])
plt.scatter(datos[['Promedio']],datos[['Tiempo']])
plt.plot(datos[['Promedio']],datos[['Y_hat']],color = 'r')
```

```python
datos['Error'] = datos['Tiempo']-datos['Y_hat']
plt.hist(datos['Error'])
```

```python
#Coeficiente de determinación
R2 = metrics.r2_score(datos[['Tiempo']],datos[['Y_hat']])
R2
```

```python
ECM = metrics.mean_squared_error(datos[['Tiempo']],datos[['Y_hat']])
ECM
```

```python
EMA = metrics.mean_absolute_error(datos[['Tiempo']],datos[['Y_hat']])
EMA
```
